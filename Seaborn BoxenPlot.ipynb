```python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.boxenplot(x=tips["total_bill"])
```

```python
ax = sns.boxenplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.boxenplot(x="day", y="total_bill", hue="smoker",data=tips, palette="Set3")
```

```python
ax = sns.boxenplot(x="day", y="total_bill", hue="time",data=tips, linewidth=2.5)
```

```python
ax = sns.boxenplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
```

```python
iris = sns.load_dataset("iris")
ax = sns.boxenplot(data=iris, orient="h", palette="Set2")
```

```python
ax = sns.boxenplot(x="day", y="total_bill", data=tips,showfliers=False)
ax = sns.stripplot(x="day", y="total_bill", data=tips,size=4, color=".26")
```

```python
g = sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="boxen",height=4, aspect=.7)
```
