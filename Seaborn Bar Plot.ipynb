```python
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
ax = sns.barplot(x="tip", y="day", data=tips)
```

```python
ax = sns.barplot(x="time", y="tip", data=tips,
                 order=["Dinner", "Lunch"])
```

```python
from numpy import median
ax = sns.barplot(x="day", y="tip", data=tips, estimator=median)
```

```python
ax = sns.barplot(x="day", y="tip", data=tips, ci=68)
```

```python
ax = sns.barplot(x="day", y="tip", data=tips, ci="sd")
```

```python
ax = sns.barplot(x="day", y="tip", data=tips, capsize=.2)
```

```python
ax = sns.barplot(x="size", y="total_bill", data=tips,palette="Blues_d")
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
ax = sns.barplot(x="day", y="total_bill", hue="weekend",data=tips, dodge=False)
```

```python
ax = sns.barplot(x="size", y="total_bill", data=tips,color="salmon", saturation=.5)
```

```python
ax = sns.barplot(x="day", y="total_bill", data=tips,linewidth=2.5, facecolor=(1, 1, 1, 0),errcolor=".2", edgecolor=".2")
```

```python
g = sns.catplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="bar",height=4, aspect=.7)
```
